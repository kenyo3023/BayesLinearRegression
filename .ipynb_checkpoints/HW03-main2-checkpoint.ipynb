```python
X, w, cov, isCov = X_simulate, blr.save_meta_iter['10']['w'], blr.save_meta_iter['10']['cov'], True

y, y_up, y_down = [],[],[]
for x in X:
    x_poly = [np.power(x, i) for i in range(blr.degree)]
    y_pred = np.dot(x_poly, w)
    y.append(y_pred)
    if isCov:
        x_poly = np.array(x_poly).reshape((1, blr.degree))
        y_var = (1 / blr.alpha) + np.dot(np.dot(x_poly, cov), x_poly.T)[0][0]
        y_up.append(y_pred + y_var)
        y_down.append(y_pred - y_var)
    else:
        y_up.append(y_pred + cov)
        y_down.append(y_pred - cov)
```
