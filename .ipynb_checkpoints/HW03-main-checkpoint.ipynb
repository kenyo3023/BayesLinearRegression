```python
import random
import math
import numpy as np
from matplotlib import pyplot as plt

state = False
def generate_Gaussian_data(mean, var):
	# Boxâ€“Muller
	global state 
	state = not state
	U = random.random()
	V = random.random()
	if state:
		X = math.sqrt(-2*math.log(U))*math.cos(2*math.pi*V)
	else:
		X = math.sqrt(-2*math.log(U))*math.sin(2*math.pi*V)
	return mean+X*math.sqrt(var)
def generate_polynomial_data(n, a, w):
	e = generate_Gaussian_data(0, a)
	x = np.random.uniform(-1,1)
	y = 0
	for i in range(n):
		y += w[i]*math.pow(x,i)
	y += e
	print(f'Add data point ({x}, {y}):')
	return x, y
def Polyline(x, coeffs, cov, isCov):
	# a n
	y,y_up,y_down = [],[],[]
	for _x in x:
		x_list = [math.pow(_x,i) for i in range(n)]
		y_predict = np.dot(x_list,coeffs)
		y.append(y_predict)
		if isCov:
			X_array = np.array(x_list).reshape((1,n))
			y_var = (1/a)+np.dot(np.dot(X_array,cov),X_array.transpose())[0][0]
			y_up.append(y_predict+y_var)
			y_down.append(y_predict-y_var)
		else:
			y_up.append(y_predict+cov)
			y_down.append(y_predict-cov)
	return y, y_up, y_down
def draw(title, x, w, cov, isCov, points_x, points_y):
	y, y_up, y_down = Polyline(x, w, cov, isCov)
	plt.plot(x, y, 'k')
	plt.plot(x, y_up, 'r')
	plt.plot(x, y_down, 'r')
	if isCov:
		plt.scatter(points_x, points_y, color='b')
	plt.title(title)
	plt.show()
```
